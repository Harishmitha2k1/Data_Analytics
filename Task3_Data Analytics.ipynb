## Visualization using Histogram(Using train dataset)

#### Load the dataset 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the training dataset 
train_df = pd.read_csv("titanic_train.csv")
```

```python
plt.subplot(231)  # 2x3 grid, 1st subplot
sns.histplot(data=train_df, x="Age", bins=20, kde=True)
plt.title("Age Histogram")
plt.show()
```

```python
plt.subplot(232)  # 2x3 grid, 2nd subplot
sns.histplot(data=train_df, x="Fare", bins=30, kde=True)
plt.title("Fare Histogram")
plt.show()
```

```python
plt.subplot(233)  # 2x3 grid, 3rd subplot
sns.histplot(data=train_df, x="Pclass", discrete=True)
plt.title("Pclass Histogram")
plt.show()
```

```python
plt.subplot(234)  # 2x3 grid, 4th subplot
sns.histplot(data=train_df, x="SibSp", discrete=True)
plt.title("SibSp Histogram")
plt.show()
```

```python
plt.subplot(235)  # 2x3 grid, 5th subplot
sns.histplot(data=train_df, x="Parch", discrete=True)
plt.title("Parch Histogram")
plt.show()
```

```python
plt.subplot(236)  # 2x3 grid, 6th subplot
sns.histplot(data=train_df, x="Survived", discrete=True)
plt.title("Survived Histogram")
plt.show()
```
